### Origin region and Destination Region
A categorical attribute representing the region for the origin and destination of the
journey respectively.
Methodology to find error:
* Find the mean values of Longitude and Latitide group by region
* Use haversine function which uses the radius of earth to find difference between 2 coordinates
* Check which distance is the smallest out of all the means.

```python
longitude_mean_df = uber_dirty_df.groupby("Origin Region")["Origin Longitude"].describe()
```

```python
longitude_mean_df = longitude_mean_df.iloc[:,0:2]
```

```python
latitude_mean_df = uber_dirty_df.groupby("Origin Region")["Origin Latitude"].describe()
latitude_mean_df = latitude_mean_df.iloc[:,0:2]
```

```python
# Merging these 2 tables into single
coordinates_mean = pd.merge(longitude_mean_df, latitude_mean_df, on=['Origin Region'])
coordinates_mean.drop("count_x", axis=1, inplace= True)
coordinates_mean.drop("count_y", axis=1, inplace= True)
coordinates_mean
```

```python
def coordinate_distance(la1, lo1, la2, lo2):
    # approximate radius of earth in km
    R = 6378.0

    lat1 = radians(la1)
    lon1 = radians(lo1)
    lat2 = radians(la2)
    lon2 = radians(lo2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
for index2, a in uber_dirty_df[["Origin Region", "Origin Latitude", "Origin Longitude"]].iterrows():
    first = True
    region = 0
    smallest_distance = 0
    for index, row in coordinates_mean.iterrows():
        distance = coordinate_distance(row[1], row[0], a[1], a[2])
        if first or (distance < smallest_distance):
            smallest_distance = distance
            region = index
            first = False
    region = float(region)
    if a[0] != region:
        print("Given region :", a[0], "Nearest region :", region) 
        uber_dirty_df.set_value(index2, 'Origin Region', region)
      
```

```python
# Function to find the difference between departure and arrival time
def get_sec(time):
    h, m, s = time.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)
```

```python
# Running the function over the 
for index2, a in uber_dirty_df[["Departure Time", "Arrival Time", "Travel Time(s)"]].iterrows():
    time = get_sec(a[1]) - get_sec(a[0])
    if (int(a[2]) != int(time)):
        print("Given time :",int(a[2])," Calculated time :", int(time))

        if abs(int(a[2])) == abs(int(time)):
            uber_dirty_df.set_value(index2, 'Departure Time', a[1])
            uber_dirty_df.set_value(index2, 'Arrival Time', a[0])
```

```python
# Function which returns the corrected distance and time as per the following assumptions.
# if calculated distance == given distance, calculated time == given time, return given values
# if calculated time = given time, calculated distance != given distance, return calculated distance
# There will not be any scenario where calculated time != given time and calculated distance != given distance
# since each row has just one error
# if calculated distance = given distance, calculated time != given time, then check if given time is present in the time 
# list and return that or else return the minimum value in the time list
def find_min(time_list, dist_list, t_given, d_given):   
    if len(time_list) == 0:
        return t_given, d_given
    else:
        min_time = min(time_list)
        min_dist = min(dist_list)
        
        for i in range(len(dist_list)):
            if time_list[i] == t_given:
                if dist_list[i] == d_given:
                    return t_given, d_given
                if dist_list[i] != d_given:
                    return t_given, dist_list[i]
            elif dist_list[i] == d_given:
                if time_list[i] != t_given:
                    if t_given in time_list:
                        return t_given, d_given
                    else:
                        return min_time, d_given
            else:
                continue
```

```python
#for index, row in uber_dirty_df[uber_dirty_df["Unnamed: 0"] == "ID3362492067"].iterrows():
#for index, row in uber_dirty_df[uber_dirty_df["Unnamed: 0"] == "ID1544421395"].iterrows():
for index, row in uber_dirty_df.iterrows():
    print(index, row[0])
    origin_lat = row[4]
    origin_long = row[5]
    dest_lat = row[6]
    dest_lon = row[7]
    src = nodes[nodes['lat'] == origin_lat][nodes['lon'] == origin_long]['F']
    if (len(src.tolist()) == 0):
        continue
    src = int(nodes[nodes['lat'] == origin_lat][nodes['lon'] == origin_long]['F'])
    des = nodes[nodes['lat'] == dest_lat][nodes['lon'] == dest_lon]['F']
    if (len(des.tolist()) == 0):
        continue
    des = int(des)
    paths = nx.all_shortest_paths(G ,source=src, target=des, weight='distance(m)')
    
    path = list(paths)
    
    time_list = []
    dist_list = []
    for each in path:
        tt = 0
        dist = 0
        for i in range(len(each)- 1):
            tt = tt + (edges[((edges.u==each[i]) & (edges.v== each[i+1])) | ((edges.v==each[i]) & (edges.u== each[i+1]))]['distance(m)'].values[0] * 1.0 * 3600) / (edges[((edges.u==each[i]) & (edges.v== each[i+1])) | ((edges.v==each[i]) & (edges.u== each[i+1]))]['speed(km/h)'].values[0] * 1000)
            dist = dist + (edges[((edges.u==each[i]) & (edges.v== each[i+1])) | ((edges.v==each[i]) & (edges.u== each[i+1]))]['distance(m)'].values[0])
        time_list.append(int(tt))
        dist_list.append(int(dist))
    t,d = find_min(time_list, dist_list, int(row[11]), int(row[8]))
    if (t != int(row[11])):
        print("Corrected time row with ID :", row[0])
        uber_dirty_df.set_value(index, 'Travel Time(s)', t)
    if (d != int(row[8])):
        print("Corrected time row with ID :", row[0])
        uber_dirty_df.set_value(index, 'Journey Distance(m)', d)
        
```

### Departure Date
Date of the departure. We know that the price is different on weekdays compared to weekends.

```python
def validate(date_text):
    # Invalid dates scenario will be handled by exception
    try:
        date = datetime.strptime(date_text, '%Y-%m-%d')
    except ValueError:
        date_split = date_text.split("-")
        # Splitting date to extract month, year and date
        year = int(date_split[0])
        month = int(date_split[1])
        day = int(date_split[2])
        
        # When month is greater than 12, it is observed that month and date are swapped. So reversing it
        if(month > 12):
            month2 = month
            month = day
            day = month2
        # In case of invalid date due to day, date is changed to 1st of next month
        elif (day >= 30):
            day = 1
            month += 1
        new_date = str(year)+ "-" + str('%02d' %month) + "-" + str('%02d' %day)
        print(date_text + " -- Changed to --> " + new_date) 
        return new_date
    else:
        return date_text
        
```

```python
# Making the missing data in the same format to predict Fare values
uber = uber_missing_backup.copy()

uber['Week'] = 0

uber['Departure Date'] = uber['Departure Date'].apply(validate)
for index, row in uber.iterrows():
    week_number = datetime.strptime(row[9], '%Y-%m-%d').weekday()
    if week_number > 5:
        uber.set_value(index, 'Week', 0)
    else:
        uber.set_value(index, 'Week', 1)
        
FMT = '%H:%M:%S'
tdelta1 = datetime.strptime("06:00:00", FMT)
tdelta2 = datetime.strptime("11:59:59", FMT)
tdelta3 = datetime.strptime("12:00:00", FMT)
tdelta4 = datetime.strptime("20:59:59", FMT)
tdelta5 = datetime.strptime("21:00:00", FMT)
tdelta6 = datetime.strptime("05:59:59", FMT)

uber['Time'] = 0
for index, row in uber.iterrows():
    time = datetime.strptime(row[10], FMT)
    if tdelta1 <= time:
        if time <= tdelta2:
            uber.set_value(index, 'Time', 0)
    if tdelta3 <= time:
        if time <= tdelta4:
            uber.set_value(index, 'Time', 1)
    if tdelta5 <= time:
        uber.set_value(index, 'Time', 2)
    if time < tdelta1:
        uber.set_value(index, 'Time', 2)
 
data2 = pd.get_dummies(uber['Time'], prefix='Time-Slot')
uber = pd.concat([uber, data2], axis = 1)

uber = uber.select_dtypes(['float64', 'int64', 'uint8'])

uber.drop('Time',axis=1, inplace= True)
uber.drop('Destination Latitude',axis=1, inplace= True)
uber.drop('Destination Longitude',axis=1, inplace= True)
uber.drop('Origin Latitude',axis=1, inplace= True)
uber.drop('Origin Longitude',axis=1, inplace= True)

uber.rename( columns={'Time-Slot_0.0':'Time-Slot_0'}, inplace=True )
uber.rename( columns={'Time-Slot_1.0':'Time-Slot_1'}, inplace=True )
uber.rename( columns={'Time-Slot_2.0':'Time-Slot_2'}, inplace=True )
uber['ID'] = uber_missing_backup["ID"]
```
